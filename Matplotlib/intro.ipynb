```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7,8]
y=[50,51,48,47,46,44,42,23]
plt.plot(x, y, color='red')
```

```python
x = ['Category A', 'Category B', 'Category C', 'Category D']
y = [25, 30, 15, 30]

# Plotting the pie chart
plt.figure(figsize=(8, 8))  # Set the figure size (optional)
plt.pie(y, labels=x, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Set aspect ratio to be equal, ensuring that the pie is drawn as a circle.
plt.axis('equal')

# Add a title (optional)
plt.title('Pie Chart Example')

# Show the plot
plt.show()
```
