```python
import pandas as pd 
import numpy as np 
```

```python
read_csv=('Data-cleaning-for-beginners-using-pandas.csv')
df=pd.read_csv(read_csv)
```

```python
df
```

```python
df.drop(columns='Index',inplace=True)
```

```python
df
```

```python
df.dropna(how='any',axis=0,inplace=True)
df
```

```python
df.replace(-1.0,1.0,inplace=True)
```

```python
df
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['Age']=df['Age'].astype(int)
```

```python
df
```

```python
df.describe()
```

```python
df['Easy Apply'].replace(-1, value=np.nan, inplace=True)
df['Established'].replace(1,value=np.nan,inplace=True)
df
```

```python
df['Established']=df['Established'].astype('Int64')
```

```python
df
```

```python
df
```

```python
df.replace({'TRUE':'YES','-1':np.nan},inplace=True)
df
```

```python
df['Easy Apply']=df['Easy Apply'].fillna('NO')
df
```

```python
df.replace({'India,In':'India','India In':'India','New York,Ny':'New York','Australia Aus':'Australia'},inplace=True)
```

```python
df.fillna(method='ffill',inplace=True)
```

```python
df
```

```python
df.set_index('Location',inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
grouped_df=df.groupby('Location')
```

```python
for Location,Location_df in grouped_df:
     print(Location)
     print(Location_df)
```

```python
df1=grouped_df.get_group(name='New York')
```
