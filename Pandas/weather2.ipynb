```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df=pd.read_csv('weather.csv',parse_dates=['day'])
df
```

```python
df.set_index('day',inplace=True)
df
```

```python
df.fillna(0)
```

```python
newdf=df.fillna({'temperature':0,'windspeed':0,'event':'no event'})
newdf
```

```python
newdf=df.fillna(method='ffill',axis=0,)  # Can also use bfill (backward fill method) #axis can also be passed 
newdf
```

```python
newdf=df.interpolate(method='time')
newdf
```

```python
newdf=df.dropna(how='all')
newdf
```

```python
dt=pd.date_range('01-01-2017','01-11-2017')
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```

```python
finaldf=df.fillna(method='ffill')
finaldf
```
