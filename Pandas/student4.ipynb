# Merge() Function

```python
df1=pd.DataFrame({'Roll No':[1,2,3,4,5],'Physics':[34,67,78,12,34]})
df1
```

```python
df2=pd.DataFrame({'Roll No':[1,2,3,4,5],'chemistry':[31,62,18,22,64]})
df2
```

```python
pd .merge(df1,df2,on='Roll No')
```

```python
pd .merge(df2,df1,on='Roll No')
```

```python
pd.merge(df1,df2)   #on- by default intersecting column is taken
```

```python
df3 = pd.DataFrame({'Roll No.':[1,2,3,6,7], 'Physics':[34,67,34,89,12]})
df4 = pd.DataFrame({'Roll No.':[1,2,3,8,9], 'Chemistry':[34,67,34,89,12]})
```

```python
pd.merge(df3, df4)
```

```python
pd.merge(df3, df4, how = 'left') # will take all the roll numbers present in the left dataframe
```

```python
pd.merge(df3, df4, how = 'right')  # will take all the roll numbers present in the right dataframe
```

```python
pd.merge(df3, df4, how = 'outer')  # will take all the roll numbers
```
